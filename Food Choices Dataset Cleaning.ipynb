# Import Library

```python
#For basic operations.
import numpy as np
import pandas as pd

#For data visualization
import matplotlib.pyplot as plt
import seaborn as sns
```

# Read The Dataset

```python
#Read the dataset.
data = pd.read_csv('food_coded.csv')

#Check the shape of the dataset
data.shape
```

# Check The Head Of The Dataset

```python
#Check the head of the dataset
data.head()
```

# Check Missing Values

```python
#Check missing values.
data.isnull().sum()
```

```python
#Impute misssing values.
data['GPA'] = data['GPA'].fillna(data['GPA'].mode()[0])
data['calories_day'] = data['calories_day'].fillna(data['calories_day'].mean())
data['type_sports'] = data['type_sports'].fillna(data['type_sports'].mode()[0])
data['weight'] = data['weight'].fillna(data['weight'].mode()[0])

#Check the missing value again.
data.isnull().sum()
```

# Check For Outliers

```python
#Check for outliers.
plt.rcParams['figure.figsize'] = (12,3)
plt.style.use('fivethirtyeight')

plt.subplot(1,3,1)
sns.boxplot(data['calories_day'])
plt.show()
```
